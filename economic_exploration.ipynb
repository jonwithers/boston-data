```python
import numpy as np
import pandas as pd
```

```python
file = "economic-indicators.csv"
```

```python
data = pd.read_csv(file)
```

```python
data = data[data['logan_passengers'] != 0]
```

The dataframe includes one line of all zero values. The above code removes it.

```python
logan_passengers = data.logan_passengers
logan_passengers_1000s = logan_passengers/10000
data = data.assign(logan_passengers_1000s = logan_passengers_1000s)
```

Added a column showing Logan passengers for that month, measured in thousands. This produces similar numbers to the average daily hotel occupancy rate and allows a nice graph of the two.

```python
data[["hotel_avg_daily_rate", "logan_passengers_1000s"]].plot(kind = "line")
```

```python
np.corrcoef(data.hotel_avg_daily_rate, data.logan_passengers_1000s)[0][1]
```

Even though the above graph seems to show a correlation between passengers at Logan and average hotel occupancy for a particular month, the exact correlation coefficient is only `0.66`.

```python
data.plot(x = "hotel_avg_daily_rate", y = "logan_passengers_1000s", kind = "scatter")
```
